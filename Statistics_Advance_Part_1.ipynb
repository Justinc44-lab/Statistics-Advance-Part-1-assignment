1. What is a random variable in probability theory
>>> In probability theory, a **random variable** is a variable that takes on different numerical values based on the outcome of a random experiment. It is a function that maps each outcome of a random process to a real number. There are two types of random variables:

 1. **Discrete Random Variable**: Takes a finite or countable number of distinct values (e.g., the number of heads in 3 coin flips).
 2. **Continuous Random Variable**: Takes an infinite number of values within a given range (e.g., the height of a randomly selected person).

Random variables are used to model uncertainty and are often denoted by letters such as X, Y, or Z.

2. What are the types of random variables?
>>> There are two main types of random variables:

 1. **Discrete Random Variable**:
   - Takes a finite or countable number of distinct values.
   - Examples: The number of cars passing through a toll booth in an hour, the number of heads in a series of coin flips.
   - These variables often use probability mass functions (PMFs) to describe the likelihood of each value.

 2. **Continuous Random Variable**:
   - Can take an infinite number of values within a given range, usually within an interval of real numbers.
   - Examples: The height of a person, the time it takes to run a race.
   - These variables are described using probability density functions (PDFs), which show the probability distribution over continuous intervals.

In summary, discrete random variables deal with countable outcomes, while continuous random variables deal with uncountably many outcomes within a range.

3. What is the difference between discrete and continuous distributions?
>>> The key differences between **discrete** and **continuous distributions** are:

 1. **Nature of Values**:
   - **Discrete Distribution**: Deals with random variables that can take on a finite or countable set of distinct values (e.g., number of successes in a series of trials).
   - **Continuous Distribution**: Deals with random variables that can take any value within a continuous range (e.g., the weight of a person, time).

 2. **Probability Representation**:
   - **Discrete Distribution**: The probability of each outcome is given by a **probability mass function (PMF)**, which assigns probabilities to each specific value.
   - **Continuous Distribution**: The probability of an exact value is zero; probabilities are given by a **probability density function (PDF)**, and the probability of a range of values is determined by the area under the curve.

 3. **Examples**:
   - **Discrete Distribution**: Binomial distribution, Poisson distribution.
   - **Continuous Distribution**: Normal distribution, Uniform distribution.

In essence, discrete distributions deal with countable outcomes, while continuous distributions involve uncountable, real-numbered outcomes.

4. What are probability distribution functions (PDF)?
>>> A **Probability Distribution Function (PDF)** describes the likelihood of different outcomes for a random variable. It is used for continuous random variables.

There are two main types of PDFs:

 1. **Probability Density Function (PDF) for Continuous Random Variables**:
   - A function that gives the relative likelihood of a random variable taking on a specific value (though the probability of any exact value is zero in continuous distributions).
   - The area under the curve of the PDF over a given range represents the probability of the random variable falling within that range.
   - The total area under the PDF curve is equal to 1.

 2. **Cumulative Distribution Function (CDF)**:
   - The CDF represents the probability that the random variable takes a value less than or equal to a specific value. It is the cumulative sum (integral) of the PDF.

**Key properties**:
- For a continuous random variable \( X \), the PDF \( f(x) \) must satisfy:
  1. \( f(x) \geq 0 \) for all \( x \).
  2. The total area under the curve (integral of \( f(x) \) over all \( x \)) equals 1.

Example: The **normal distribution** has a well-known bell-shaped PDF.

In short, a PDF describes how probabilities are distributed over the possible outcomes of a continuous random variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
>>> The key difference between a **Cumulative Distribution Function (CDF)** and a **Probability Distribution Function (PDF)** is their purpose and what they represent:

 1. **Cumulative Distribution Function (CDF)**:
   - The CDF gives the cumulative probability that a random variable \( X \) takes a value less than or equal to a specific value \( x \).
   - Mathematically: \( \text{CDF}(x) = P(X \leq x) \).
   - It shows the accumulated probability up to a certain point.
   - The CDF always ranges from 0 to 1 and is a non-decreasing function.

 2. **Probability Distribution Function (PDF)**:
   - The PDF describes the relative likelihood of a continuous random variable taking a specific value (although the probability of a single value is 0 for continuous variables).
   - For continuous random variables, the PDF gives the height of the distribution at a given point.
   - The area under the curve of the PDF between two points gives the probability that the variable falls within that range.

**In short**:
- **CDF** provides the cumulative probability up to a point, while **PDF** gives the likelihood of the variable taking specific values (for continuous variables). The PDF is the derivative of the CDF.

6. What is a discrete uniform distribution?
>>> A **discrete uniform distribution** is a probability distribution where a finite number of outcomes are equally likely to occur. In other words, each outcome has the same probability.

- **Definition**: For a discrete uniform distribution, if a random variable \( X \) can take \( n \) distinct values (say, \( x_1, x_2, \dots, x_n \)), then the probability of each value is the same, given by:

  \[
  P(X = x_i) = \frac{1}{n}, \quad \text{for each } i = 1, 2, \dots, n
  \]

- **Examples**:
  - Rolling a fair die (with outcomes 1, 2, 3, 4, 5, or 6), where each face has a probability of \( \frac{1}{6} \).
  - Selecting a random card from a well-shuffled deck (assuming all cards are equally likely to be chosen).

**Key Characteristics**:
- All outcomes have equal probability.
- The distribution is discrete (countable outcomes).
- The total probability sums to 1.

In short, a discrete uniform distribution is one where every possible outcome has the same probability.

7. What are the key properties of a Bernoulli distribution?
>>> The **Bernoulli distribution** is a discrete probability distribution that models a random experiment with exactly two possible outcomes: success (usually denoted as 1) or failure (denoted as 0). It is often used to represent binary or yes/no situations.

### Key Properties of a Bernoulli Distribution:

 1. **Two Possible Outcomes**:
   - The random variable \( X \) can take one of two values: 0 (failure) or 1 (success).

 2. **Probability of Success**:
   - Let \( p \) be the probability of success (i.e., \( X = 1 \)).
   - The probability of failure is \( 1 - p \) (i.e., \( X = 0 \)).

 3. **Probability Mass Function (PMF)**:
   - The PMF of a Bernoulli random variable \( X \) is:
     \[
     P(X = 1) = p \quad \text{and} \quad P(X = 0) = 1 - p
     \]
   - where \( 0 \leq p \leq 1 \).

 4. **Mean (Expected Value)**:
   - The expected value \( E[X] \) is equal to the probability of success:
     \[
     E[X] = p
     \]

 5. **Variance**:
   - The variance \( \text{Var}(X) \) is given by:
     \[
     \text{Var}(X) = p(1 - p)
     \]

 6. **Support**:
   - The possible outcomes are \( 0 \) and \( 1 \).

### Example:
- A coin flip where the probability of getting heads (success) is \( p = 0.5 \), and the probability of getting tails (failure) is \( 1 - 0.5 = 0.5 \).

In short, a Bernoulli distribution models a single trial of a binary outcome with a success probability \( p \) and failure probability \( 1 - p \).

8. What is the binomial distribution, and how is it used in probability?
>>> The **binomial distribution** is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

### Key Features of the Binomial Distribution:

1. **Number of Trials (n)**:
   - The distribution models \( n \) independent trials or experiments.
   
2. **Probability of Success (p)**:
   - Each trial has a probability \( p \) of success (and \( 1 - p \) for failure).

3. **Random Variable (X)**:
   - The random variable \( X \) represents the number of successes in \( n \) trials.

4. **Probability Mass Function (PMF)**:
   - The probability of exactly \( k \) successes (where \( k \) is between 0 and \( n \)) is given by the formula:
     \[
     P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
     \]
     where \( \binom{n}{k} \) is the binomial coefficient, which gives the number of ways to choose \( k \) successes out of \( n \) trials.

### Example:
- Tossing a fair coin 5 times (n = 5), and finding the probability of getting exactly 3 heads (k = 3), where \( p = 0.5 \).

### Applications in Probability:
- The binomial distribution is used to calculate the probability of a given number of successes in scenarios like:
  - Quality control (e.g., defective items in a batch).
  - Voting outcomes (e.g., number of votes for a candidate).
  - Medical trials (e.g., number of patients who respond to treatment).

In short, the binomial distribution helps model scenarios with a fixed number of independent trials, where each trial has two possible outcomes and the same probability of success.

9. What is the Poisson distribution and where is it applied?
>>> The **Poisson distribution** is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space, given that the events happen independently and at a constant average rate.

### Key Features of the Poisson Distribution:
 1. **Events occur randomly**:
   - The distribution assumes that events happen independently and are spread out randomly over time or space.

 2. **Parameter (λ)**:
   - The distribution is characterized by a single parameter \( \lambda \) (lambda), which represents the average rate of occurrences (mean number of events) in the given interval.

 3. **Probability Mass Function (PMF)**:
   - The probability of observing exactly \( k \) events in an interval is given by:
     \[
     P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
     \]
     where \( k = 0, 1, 2, \dots \) and \( \lambda > 0 \).

### Applications of Poisson Distribution:
- **Modeling rare events** that occur over time or space, where the event rate is constant:
  - **Call centers**: The number of calls received in an hour.
  - **Traffic flow**: The number of cars passing a checkpoint in a given period.
  - **Queuing systems**: The number of customers arriving at a service point.
  - **Natural events**: The number of earthquakes occurring in a region within a year.
  - **Website traffic**: The number of hits to a website in a given time period.

### Summary:
The Poisson distribution is used to model the occurrence of rare events over time or space, where these events are independent and happen at a constant average rate. It is especially useful in cases where events happen infrequently.

10. What is a continuous uniform distribution?
>>> A **continuous uniform distribution** is a probability distribution where all outcomes within a certain interval are equally likely. In other words, every value in a given range has the same probability of occurring.

### Key Features of the Continuous Uniform Distribution:

 1. **Range (a, b)**:
   - The random variable \( X \) can take any value between \( a \) and \( b \), where \( a \) is the minimum value and \( b \) is the maximum value of the interval.

 2. **Probability Density Function (PDF)**:
   - The PDF of a continuous uniform distribution is constant over the interval \( [a, b] \) and is given by:
     \[
     f(x) = \frac{1}{b - a}, \quad \text{for } a \leq x \leq b
     \]
   - The total area under the PDF curve equals 1, representing the total probability.

 3. **Mean (Expected Value)**:
   - The mean of the distribution is the midpoint of the interval:
     \[
     E[X] = \frac{a + b}{2}
     \]

 4. **Variance**:
   - The variance of the distribution is:
     \[
     \text{Var}(X) = \frac{(b - a)^2}{12}
     \]

### Example:
- If you randomly pick a number between 0 and 10, the probability of picking any number in that range (e.g., 2, 5, 8) is the same.

### Applications:
- **Random sampling**: When a value is randomly selected from a continuous range with equal probability (e.g., selecting a random time between 2 and 5 PM).
- **Simulation**: Often used in simulations or modeling where outcomes are evenly distributed over a given interval.

In short, the continuous uniform distribution describes situations where all values within a specific range are equally likely to occur.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
>>> ### **Central Limit Theorem (CLT):**

The **Central Limit Theorem** (CLT) states that, regardless of the original distribution of a population, the sampling distribution of the sample mean will approach a **normal distribution** as the sample size \( n \) increases, provided that the samples are independent and identically distributed (i.i.d.).

### Key Points:

**A. Sampling Distribution of the Mean**  
   - When you take multiple random samples from any population and calculate the mean of each sample, the distribution of those sample means will tend to be normal as the sample size increases.

**B. Sample Size**  
   - The larger the sample size \( n \), the closer the sampling distribution of the mean will be to a normal distribution, even if the population itself is not normally distributed.

**C. Mean and Standard Deviation of Sampling Distribution**  
   - The mean of the sampling distribution will be the same as the population mean \( \mu \).
   - The standard deviation of the sampling distribution (called the **standard error**) is given by:
     \[
     \text{Standard Error} = \frac{\sigma}{\sqrt{n}}
     \]
     where \( \sigma \) is the population standard deviation and \( n \) is the sample size.

### **Importance of CLT in Statistics:**

**D. Facilitates Inference**  
   - The CLT allows statisticians to use the normal distribution for hypothesis testing and confidence intervals, even if the original population is not normally distributed, as long as the sample size is sufficiently large.

**E. Basis for Many Statistical Methods**  
   - It is the foundation of many statistical methods, including estimation, regression, and significance testing, because it justifies approximating sample means with a normal distribution.

**F. Practical Application**  
   - In practice, CLT enables analysis of sample data without needing to know the underlying distribution of the population, as long as the sample size is large enough.

### In short, the Central Limit Theorem is critical because it ensures that sample means follow a normal distribution for large sample sizes, making it possible to apply statistical techniques based on normality even when the original data is not normally distributed.

15.  What is the application of Z statistics in hypothesis testing?
>>> ### **Application of Z-Statistics in Hypothesis Testing:**

**A. Standardization of Data**  
   - **Z-statistics** are used to standardize sample data by converting it into **Z-scores**, which represent how many standard deviations a sample mean is from the population mean. This is done using the formula:
     \[
     Z = \frac{\overline{X} - \mu}{\sigma / \sqrt{n}}
     \]
     where:
     - \( \overline{X} \) is the sample mean,
     - \( \mu \) is the population mean,
     - \( \sigma \) is the population standard deviation,
     - \( n \) is the sample size.

**B. Hypothesis Testing Steps**  
   1. **Null and Alternative Hypotheses**: The null hypothesis (\( H_0 \)) typically states that there is no difference, while the alternative hypothesis (\( H_a \)) proposes a difference.
   2. **Z-statistic Calculation**: Calculate the Z-statistic based on the sample data.
   3. **Determine Critical Value**: Find the critical Z-value(s) corresponding to the desired significance level (\( \alpha \)) from the Z-table.
   4. **Decision Rule**: Compare the calculated Z-statistic to the critical value(s):
      - If the Z-statistic falls in the rejection region (outside the critical value), reject \( H_0 \).
      - If the Z-statistic falls in the acceptance region, fail to reject \( H_0 \).

**C. When to Use Z-Statistics**  
   - Z-statistics are used when:
     - The sample size is large (\( n \geq 30 \)) or the population standard deviation \( \sigma \) is known.
     - The underlying distribution of the population is approximately normal.

**D. Applications in Hypothesis Testing**  
   - **One-sample Z-test**: Used to compare the sample mean to a known population mean.
   - **Two-sample Z-test**: Used to compare the means of two independent samples (when population variances are known).
   - **Proportion Z-test**: Used to test hypotheses about population proportions.

### In short, Z-statistics are used in hypothesis testing to standardize data, calculate test statistics, and make decisions about population parameters, especially when the population standard deviation is known and sample sizes are large.

16. How do you calculate a Z-score, and what does it represent?
>>> ### **Calculating a Z-Score:**

A **Z-score** represents how many standard deviations a data point is from the mean of the distribution. It is calculated using the following formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- **X** is the individual data point,
- **μ** is the mean of the population,
- **σ** is the standard deviation of the population.

### **What Does a Z-Score Represent?**

**A. Relative Position**  
   - A **positive Z-score** indicates that the data point is above the mean.
   - A **negative Z-score** indicates that the data point is below the mean.
   
**B. Standard Deviations**  
   - The Z-score tells you how far away a data point is from the mean in terms of **standard deviations**.
   - For example, a Z-score of +2 means the data point is 2 standard deviations above the mean, while a Z-score of -1 means the data point is 1 standard deviation below the mean.

**C. Comparison Across Different Distributions**  
   - Z-scores allow comparison of data points from different distributions, even if those distributions have different means and standard deviations, by standardizing the values.

### In short, a Z-score measures how far a data point is from the mean in terms of standard deviations, helping to compare values across different distributions.

17. What are point estimates and interval estimates in statistics?
>>> ### **Point Estimates and Interval Estimates in Statistics:**

**A. Point Estimate:**
   - A **point estimate** is a single value used to estimate an unknown population parameter (such as the mean or proportion).
   - It is derived from sample data and serves as a best guess for the population parameter.
   - Example: The sample mean (\( \overline{X} \)) is a point estimate of the population mean (\( \mu \)).

**B. Interval Estimate:**
   - An **interval estimate** provides a range of values within which the true population parameter is likely to fall, along with a specified level of confidence.
   - This range is typically expressed as a **confidence interval** (e.g., \( \mu \in [L, U] \)), where \( L \) and \( U \) are the lower and upper bounds.
   - Example: A 95% confidence interval for the population mean might be \( [5.2, 6.8] \), indicating that we are 95% confident the true mean lies within this range.

### **Key Differences:**

**C. Point Estimate:**
   - Provides a single value.
   - Less precise, as it does not account for variability.

**D. Interval Estimate:**
   - Provides a range of possible values.
   - More informative, as it quantifies uncertainty and provides a confidence level.

### In short, **point estimates** provide a single best guess of a parameter, while **interval estimates** give a range of values that likely contain the parameter, offering more precision and confidence.

   - Confidence intervals give more information than point estimates, as they provide both an estimate and a measure of precision. A narrower interval indicates a more precise estimate, while a wider interval suggests more uncertainty.

   - Confidence intervals are used in hypothesis testing to assess whether a parameter lies within a given range. For instance, if a confidence interval for a mean does not include 0, it suggests a statistically significant result.

**D. Decision-Making**  
   - In real-world applications, confidence intervals help guide decisions by giving an idea of the likely range of outcomes and the level of certainty associated with them.

### In short, **confidence intervals** are significant in statistical analysis because they provide a range of plausible values for a parameter, help quantify uncertainty, and support decision-making by offering both estimate and precision.

19. What is the relationship between a Z-score and a confidence interval?
>>> ### **Relationship Between Z-Score and Confidence Interval:**

**A. Z-Score in Confidence Intervals**  
   - The **Z-score** is used to determine the **critical value** when constructing a confidence interval, especially when the population standard deviation is known and the sample size is large.
   - The Z-score represents the number of standard deviations a data point is from the mean, and it helps define the margin of error around the sample estimate.

**B. Confidence Interval Formula**  
   - A confidence interval is often constructed using the formula:
     \[
     \text{Confidence Interval} = \overline{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
     \]
     Where:
     - \( \overline{X} \) is the sample mean,
     - \( Z_{\alpha/2} \) is the critical Z-value corresponding to the desired confidence level (e.g., for a 95% confidence level, \( Z_{\alpha/2} \approx 1.96 \)),
     - \( \sigma \) is the population standard deviation,
     - \( n \) is the sample size.

**C. Z-Score as a Confidence Level Indicator**  
   - The Z-score value corresponds to a specific confidence level:
     - A **Z-score of 1.96** corresponds to a 95% confidence interval, meaning 95% of the sample means will fall within this range.
     - A **Z-score of 1.64** corresponds to a 90% confidence level.

### In short, the **Z-score** is used to determine the critical value for constructing a **confidence interval**, helping define the range within which the true population parameter is likely to fall, based on a desired confidence level.

21. What are the assumptions for applying the Central Limit Theorem?
>>> ### **Assumptions for Applying the Central Limit Theorem (CLT):**

**A. Random Sampling**  
   - The data must be collected through **random sampling**, ensuring that each observation is independent and representative of the population.

**B. Sample Size**  
   - The sample size \( n \) must be sufficiently large. Generally, a sample size of **n ≥ 30** is considered large enough for the CLT to apply, although larger sample sizes are preferred if the population distribution is highly skewed or has outliers.

   - The observations within the sample must be **independent** of each other. This means that the value of one observation should not influence the value of another.

   - The population does not need to be normally distributed. However, for smaller sample sizes, the population should be approximately normal. ### In short, the **Central Limit Theorem** assumes random sampling, sufficient sample size, independence of observations, and, for smaller sample sizes, an approximately normal population distribution. For larger sample sizes, the population distribution can be non-normal.

22. What is the concept of expected value in a probability distribution?
>>> ### **Concept of Expected Value in a Probability Distribution:**

**A. Definition**  
   - The **expected value** (often denoted as \(E[X]\)) is the **long-run average** or **mean** of a random variable in a probability distribution. It represents the center or "balance point" of the distribution.

**B. Formula for Discrete Random Variables**  
   - For a **discrete random variable**, the expected value is calculated as:
     \[
     E[X] = \sum_{i=1}^{n} x_i \cdot P(x_i)
     \]
     Where:
     - \(x_i\) represents each possible value of the random variable,
     - \(P(x_i)\) is the probability of \(x_i\) occurring.

**C. Formula for Continuous Random Variables**  
   - For a **continuous random variable**, the expected value is calculated as:
     \[
     E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
     \]
     Where:
     - \(f(x)\) is the probability density function (PDF) of the continuous random variable.

**D. Interpretation**  
   - The expected value gives the **average outcome** if an experiment is repeated many times. It is not necessarily a value that will occur in any specific trial, but a theoretical average over the long run.

### In short, the **expected value** is the weighted average of all possible outcomes in a probability distribution, giving a central measure of what you can expect on average from a random variable.

23. How does a probability distribution relate to the expected outcome of a random variable?
>>> ### **Relationship Between Probability Distribution and Expected Outcome of a Random Variable:**

**A. Probability Distribution**  
   - A **probability distribution** describes how the probabilities are distributed across the possible values of a random variable. It provides the likelihood of each possible outcome occurring.

**B. Expected Outcome (Expected Value)**  
   - The **expected value** is the **weighted average** of all possible outcomes of a random variable, with the weights being the probabilities from the probability distribution.
   - It represents the **long-term average** or **center** of the distribution, indicating what you would expect on average if the experiment were repeated many times.

**C. Formula for Expected Value**  
   - For a **discrete random variable**, the expected value \(E[X]\) is calculated using the probability distribution as:
     \[
     E[X] = \sum_{i=1}^{n} x_i \cdot P(x_i)
     \]
     Where \(x_i\) are the possible values and \(P(x_i)\) are their corresponding probabilities.

**D. Interpretation**  
   - The probability distribution tells you the chances of different outcomes, while the expected value gives you the **average outcome** based on these chances.

### In short, the **probability distribution** defines the likelihood of each outcome, and the **expected value** calculates the average outcome based on those probabilities, giving an overall sense of what to expect in the long run.

```python
#19  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, population_mean, population_std, alpha=0.05):
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 3: Calculate the P-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed p-value

    # Step 4: Decision Rule
    if p_value < alpha:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    # Output results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    # Plotting the standard normal distribution and marking the Z-statistic
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = stats.norm.pdf(x_vals, 0, 1)

    plt.plot(x_vals, y_vals, label='Standard Normal Distribution')
    plt.axvline(z_statistic, color='red', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')
    plt.fill_between(x_vals, 0, y_vals, where=(x_vals >= z_statistic), color='red', alpha=0.3)

    plt.title(f"Z-Test: {decision}")
    plt.xlabel('Z-value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example data and parameters
sample_data = np.array([55, 60, 58, 62, 59, 57, 61, 63, 64, 60])  # Example sample data
population_mean = 60  # Population mean (null hypothesis)
population_std = 5   # Known population standard deviation

# Perform the Z-test
z_test(sample_data, population_mean, population_std, alpha=0.05)
```

```python
#20  Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
    - data: The sample data (list or numpy array).
    - confidence_level: The confidence level for the interval (default is 0.95).

    Returns:
    - tuple: (lower_bound, upper_bound) of the confidence interval.
    """
    # Sample mean and sample standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (use ddof=1 for unbiased estimate)
    n = len(data)  # Sample size

    # Standard error of the mean
    se = sample_std / np.sqrt(n)

    # Find the t-score for the confidence level (use t-distribution for small sample sizes)
    confidence_interval = stats.t.interval(confidence_level, df=n-1, loc=sample_mean, scale=se)

    return confidence_interval

# Example dataset
data = np.array([55, 60, 58, 62, 59, 57, 61, 63, 64, 60])

# Calculate the confidence interval
lower, upper = confidence_interval(data, confidence_level=0.95)

# Print the result
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Plotting the data with the confidence interval
plt.figure(figsize=(8, 5))
plt.hist(data, bins=6, alpha=0.75, color='skyblue', edgecolor='black', label='Data')
plt.axvline(lower, color='red', linestyle='--', label=f'Lower Bound: {lower:.2f}')
plt.axvline(upper, color='green', linestyle='--', label=f'Upper Bound: {upper:.2f}')
plt.axvline(np.mean(data), color='black', linestyle='-', label=f'Mean: {np.mean(data):.2f}')
plt.legend()
plt.title('95% Confidence Interval')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#24  Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

def simulate_uniform_distribution(a, b, n_samples):
    """
    Simulate a random variable using a continuous uniform distribution
    and calculate its expected value (mean).

    Parameters:
    - a: The lower bound of the uniform distribution.
    - b: The upper bound of the uniform distribution.
    - n_samples: The number of random samples to simulate.

    Returns:
    - The calculated sample mean (which approximates the expected value).
    """
    # Simulate random samples from a uniform distribution
    samples = np.random.uniform(a, b, n_samples)

    # Calculate the sample mean (this should approximate the expected value)
    sample_mean = np.mean(samples)

    # Calculate the expected value using the formula E[X] = (a + b) / 2
    expected_value = (a + b) / 2

    return sample_mean, expected_value

# Parameters for the uniform distribution
a = 2  # Lower bound
b = 10  # Upper bound
n_samples = 10000  # Number of random samples to simulate

# Simulate the distribution and calculate the expected value
sample_mean, expected_value = simulate_uniform_distribution(a, b, n_samples)

# Print results
print(f"Sample Mean (approximated expected value): {sample_mean:.4f}")
print(f"Expected Value (theoretical): {expected_value:.4f}")
```

```python
#25 Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    - data1: First dataset (list or numpy array).
    - data2: Second dataset (list or numpy array).
    """
    # Calculate standard deviations
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (ddof=1)
    std_dev2 = np.std(data2, ddof=1)

    # Print the standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.4f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.4f}")

    # Visualize the datasets using boxplots
    plt.figure(figsize=(10, 6))

    # Create box plots for the two datasets
    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'], patch_artist=True,
                boxprops=dict(facecolor='skyblue', color='blue'),
                whiskerprops=dict(color='blue'), capprops=dict(color='blue'))

    # Title and labels
    plt.title('Comparison of Standard Deviations Between Two Datasets')
    plt.ylabel('Values')
    plt.grid(True)

    # Show the plot
    plt.show()

# Example datasets
data1 = np.random.normal(loc=10, scale=5, size=1000)  # Mean = 10, Std Dev = 5
data2 = np.random.normal(loc=10, scale=2, size=1000)   # Mean = 10, Std Dev = 2

# Compare the standard deviations and visualize
compare_standard_deviations(data1, data2)
```

```python
#26  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a given dataset.

    Parameters:
    - data: The dataset (list or numpy array).

    Returns:
    - range_value: The range of the dataset.
    - iqr_value: The interquartile range (IQR) of the dataset.
    """
    # Calculate the range (max - min)
    range_value = np.max(data) - np.min(data)

    # Calculate the IQR (Q3 - Q1)
    q1 = np.percentile(data, 25)  # First quartile (25th percentile)
    q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
    iqr_value = q3 - q1

    return range_value, iqr_value

# Generate a dataset from a normal distribution
mean = 50  # Mean of the normal distribution
std_dev = 10  # Standard deviation of the normal distribution
n_samples = 1000  # Number of samples

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=n_samples)

# Calculate range and IQR
range_value, iqr_value = calculate_range_and_iqr(data)

# Print the results
print(f"Range of the dataset: {range_value:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr_value:.2f}")
```

```python
#27  Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization on the dataset.

    Parameters:
    - data: The dataset (list or numpy array).

    Returns:
    - normalized_data: The dataset after Z-score normalization.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Generate a dataset from a normal distribution
mean = 50  # Mean of the normal distribution
std_dev = 15  # Standard deviation of the normal distribution
n_samples = 1000  # Number of samples

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=n_samples)

# Apply Z-score normalization
normalized_data = z_score_normalization(data)

# Plot the original and normalized datasets for comparison
plt.figure(figsize=(12, 6))

# Plot original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data (Before Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Plot normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Normalized Data (After Z-score Normalization)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()

# Print mean and std deviation of original and normalized data
print(f"Original Data Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
print(f"Normalized Data Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")
```

```python
#28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a given dataset.

    Parameters:
    - data: The dataset (list or numpy array).

    Returns:
    - skew_value: The skewness of the dataset.
    - kurtosis_value: The kurtosis of the dataset.
    """
    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (excess kurtosis, where normal distribution has kurtosis=0)
    kurtosis_value = kurtosis(data)

    return skew_value, kurtosis_value

# Generate a dataset from a normal distribution
mean = 50  # Mean of the normal distribution
std_dev = 10  # Standard deviation of the normal distribution
n_samples = 1000  # Number of samples

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=n_samples)

# Calculate skewness and kurtosis
skew_value, kurtosis_value = calculate_skewness_and_kurtosis(data)

# Print the results
print(f"Skewness of the dataset: {skew_value:.4f}")
print(f"Kurtosis of the dataset: {kurtosis_value:.4f}")
```
